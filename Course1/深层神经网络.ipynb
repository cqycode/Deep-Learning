### 深层神经网络



四层神经网络，输入层不计入

$n^{[i]}$表示第i层的结点数 $n^{[1]}=5,n^{[2]}=4,n^{[3]}=2,n^{[4]}=1$  
$a^{[i]}$表示第i层的激活函数  


#### 前向传播

$Z^{[l]}=W^{[l]}*A^{[l-1]}+b^{[l]}$

$A^{[l]}=g^{[l]}(Z^{[l]})$

#### 反向传播

$dZ^{[l]}=dA^{[l]}*g^{[l]'}(Z^{[l]})$

$dW^{[l]}=\frac1mdZ^{[l]}*A^{[l-1]T}$

$db^{[l]}=\frac1mnp.sum(dZ^{[l]},axis=1)$

$dA^{[l-1]}=W^{[l]T}*dZ^{[l]}$
