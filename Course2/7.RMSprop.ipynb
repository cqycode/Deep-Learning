### RMSprop
RMSprop全称是root mean square prop算法。

假设我们的神经网络只有两个参数：w、b。以w为横轴，b为纵轴画出cost function的等高线图见下：


目标：减少纵向的摆动，加快横向的进程

原因:很明显是因为$d_b$ > $d_w$，因此在相同的学习率α下，才造成了b更新的步长要比w大很多。

解决：对参数b使用更小的学习率$αλ_1$，而对参数w使用更大的学习率$αλ_2$，$λ_1$<$λ_2$

实现过程：
- On iteration t:
    - Compute dw,db on current mini-batch
    - $S_{dw}=\beta S_{dw}+(1-\beta) (dw)^2$
    - $S_{db}=\beta S_{db}+(1-\beta)(db)^2$
    - $w=w-\alpha \frac{dw}{\sqrt{S_{dw}}}$
    - $b=b-\alpha \frac{db}{\sqrt{S_{db}}}$


这里的$\frac{1}{\sqrt{S_{db}}}$ 就相当于前文提到的$\lambda_1$，$\frac{1}{\sqrt{S_{dw}}}$就相当于$\lambda_2$

最后，为了防止分母$\sqrt{S_{db}}$,$\sqrt{S_{dw}}$趋近于0，从而造成$\frac{dw}{\sqrt{S_{dw}}}$,$\frac{db}{\sqrt{S_{db}}}$的值过大，因此在实际应用中，将分母加上一个很小很小的值$\epsilon$，通常取$\epsilon=10^{-8}$。

修正完的公式见下：

$w:=w-\alpha \frac{dw}{\sqrt{S_{dw}}+\epsilon}$

$b:=b-\alpha \frac{db}{\sqrt{S_{db}}+\epsilon}$
